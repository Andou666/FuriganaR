```r
# 残高がゼロになるまで繰り返す
shikin = 50000
while(shikin >= 0){
    print(shikin)
    shikin <- shikin - 5080
}
```

```r
 # 同じメッセージを10回表示する
for(i in 1:10){
    print("ハロー！")
}
```

```r
# メッセージの中に回数を入れる
for(i in 1:10){
    print(paste(i,"回目のハロー！",sep=""))
}
```

```r
# 逆順で繰り返すには
for(i in 10:1){
    print(paste(i,"回目のハロー！",sep=""))
}
```

```r
# 九九の計算をしてみよう
for(x in 1:9){
    for(y in 1:9){
        print(x*y)
    }
}
```

```r
# 九九らしく表示する
for(x in 1:9){
    for(y in 1:9){
        print(paste(x,"×",y,"＝",x*y))
    }
}
```

```r
# 配列* を作って利用する
wdays <- c('月', '火', '水', '木', '金')
print(wdays[2])
```

```r
# 配列の要素を書き替える
wdays <- c('月', '火', '水', '木', '金')
wdays[2] <- "炎"
print(wdays);
```

```r
# for文を使ってインデックスを指定する
wdays <- c('月', '火', '水', '木', '金')
for(i in 1:5){
print(paste(wdays[i],"曜日",sep=""));
}
```

```r
# 単純にすべての組み合わせを並べる
team <- c('A', 'B', 'C', 'D', 'E')
for(t1 in team){
    for(t2 in team){
        print(paste(t1,' vs ',t2))
    }
}
```

```r
team <- c('A', 'B', 'C', 'D', 'E')
for(t1 in team){
    for(t2 in team){
        if(t1 != t2){
            print(paste(t1,' vs ',t2))
        }
    }
}
```

```r
# 同じ対戦組み合わせを省くには？
team <- c('A', 'B', 'C', 'D', 'E')
opps <- c('A', 'B', 'C', 'D', 'E')
for(t1 in team){
    opps <- opps[-1]
    for(t2 in opps){
        print(paste(t1,' vs ',t2))
    }
}
```
